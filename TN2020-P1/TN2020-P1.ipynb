### Αλγόριθμος παραγωγής λαβυρίνθων
```
1. Αρχικά όλα τα κελιά περιέχουν εμπόδια και η λίστα των τοίχων είναι κενή.
2. Ένα κελί επιλέγεται ως μέρος του λαβυρίνθου. Οι τοίχοι του κελιού προστίθενται στη λίστα των τοίχων.
3. Όσο η λίστα των τοίχων δεν είναι κενή:
      1. Επιλέγεται τυχαία ένας τοίχος από τη λίστα. Εαν μόνο ένα από τα (δύο) κελιά που χωρίζει ο τοίχος είναι μέρος του λαβυρίνθου, τότε:
            1. Ο τοίχος και το κελί που δεν είναι μέρος του λαβυρίνθου, γίνονται μέρος του λαβυρίνθου.
            2. Οι τοίχοι των νέων κελιών που προστέθηκαν στο λαβύρινθο προστίθενται στη λίστα των τοίχων.
      2. Ο τοίχος αφαιρείται από τη λίστα των τοίχων
```
*ως τοίχοι ενός κελιού θεωρούνται όσα απο τα 4 γειτονικά κελιά (όχι τα διαγώνια) περιέχουν εμπόδιο

````
%matplotlib inline
import numpy as np
from queue import LifoQueue
from random import shuffle, uniform
import matplotlib.pyplot as plt
import random
import time



class grid:
    def __init__(self, N, S, F, part3 = False, limits_part3 = (0,1)):   

        # variable part3 is used for the third part of the assignment 
        # if part3 = True : then all cells of the maze that are part of the labyrinth get a random cost from a specified distribution
        # if part3 = False : we dont specify the costs of the cells  
        # limits_part3 : is used for the third part of the assignment and it specifies the range of the costs, each distribution applies
        
        ## Make sure start and end are within the grid
        assert N > 2
        assert S[0] < N
        assert S[1] < N
        assert F[0] < N
        assert F[1] < N

        assert S[0] > 0
        assert S[1] > 0
        assert F[0] > 0
        assert F[1] > 0

        self.N = N
        self.part3 = part3
        ## Initialize grid with obstacles
        self.grid = np.ones((N, N), dtype=np.int32)
        self.limits_part3 = limits_part3
        ## Start and end position have no obstacles
        self.grid[S]=0
        self.grid[F]=0
        obstacle_free_points = {S, F}
        
        ### YOUR CODE HERE        
        i = 0
        self.generate_maze(N,S,F)  #create the labyrinth
        self.simplify_maze(N,S,flag = True, prob = 0.15) #if flag = True simplify the maze with a prob  
        uniform_weight, triangular_weight, beta_weight = self.distributions(N, S, part3, limits_part3) 
        self.bfs(S,F,N,i)                             
        
          

    def bfs(self, S, F, N, i):

      self.S = S
      self.F = F
      self.N = N
      self.i = i
      flag = 0 # if flag remains 0 => we need to apply recursion because we couldnt find a path
      visited = []
      queue = []
      dead_points = []
      queue.append(F)
      visited.append(F)

      while queue:
        points = queue.pop(0) #pop the first element of the list
        if points == S: #if we reached the start point, return
          flag = 1
          if self.part3 == False:   
            print('PATH FOUND')
          return
        for neighbour in self.neighbours(N,points): #for each neighbour
          if neighbour not in visited: #if we havent visited him
            if self.grid[neighbour] == 0: #if neighbour is part of the labyrinth
              visited.append(neighbour)   
              queue.append(neighbour)   #insert him in the queue and list him as visited
            else:
              dead_points.append(neighbour)
      if flag == 0:  #if we havent found a path
        print('Couldnt track a path so we apply recursion')
        make_white = dead_points.pop(i)  #pop one Blocked cell that is closer to F
        i += 3    #defines the Blocked cell to be popped
        self.grid[make_white] = 0  
        self.bfs(S, F, N, i) #recursion

    def distributions(self, N, S, part3, tuple_of_bountries):
      if part3 == True:
        a, b = tuple_of_bountries
        uniform_weight = {S : 0}   #create 3 dicts one for each distribution
        beta_weight = {S : 0}
        triangular_weight = {S : 0}
        for i in range(N):
          for j in range(N):
            if self.grid[i][j] == 0: #search all the grid for parts of the labyrinth
              if (i,j) != S:         
                if (a,b) == (0,1): #we select the right case which corresponds to the tuple given
                  uniform_weight[(i,j)] = np.random.uniform(low=np.nextafter(0.0, 1.0), high=1.0)
                  triangular_weight[(i,j)] = np.random.triangular(left = np.nextafter(0.0, 1.0), mode = 0.5,  right = np.nextafter(1.0, 0.0))
                  beta_weight[(i,j)] = np.random.beta(a = 1, b = 1)
                elif (a,b) == (1,2):
                  uniform_weight[(i,j)] = np.random.uniform(low=np.nextafter(1.0, 2.0), high=2.0)
                  triangular_weight[(i,j)] = np.random.triangular(left = np.nextafter(1.0, 2.0), mode = 1.5,  right = np.nextafter(2.0, 1.0))
                  beta_weight[(i,j)] = np.random.beta(a = 1, b = 1) + 1
                elif (a,b) == (0, 0.1):
                  uniform_weight[(i,j)] = np.random.uniform(low = 0, high=0.1)
                  triangular_weight[(i,j)] = np.random.triangular(left = 0, mode = 0.05,  right = np.nextafter(0.1, 0.0))
                  beta_weight[(i,j)] = np.random.beta(a = 1, b = 1) / 10
                else:
                  raise TypeError('Give a tuple from the following: (0,1), (1,2), (0,0.1)')
        return uniform_weight, triangular_weight, beta_weight
      
      else: #part3 = False return empty dicts
        return 0, 0, 0

    def neighbours(self,N,F):  # (+-)1 neighbours are used in bfs function and astar_search(part2)
      neighbours = []
      x, y = F[0], F[1]
      if x in range(N) and y in range(N):
        neighbours = ([(i,j) for i,j in [(x+1,y), (x-1,y), (x, y-1), (x, y+1)] if i in range(N) and j in range(N)])
     
      return neighbours


    def simplify_maze(self,N, S, flag, prob):

      if flag == False:
        return
      if flag == True:
        for i in range(N):
          for j in range(N):
            if self.grid[i][j] == 1:
              self.grid[i][j] = np.random.choice(np.arange(0,2), p = [prob, 1-prob]) #choose 0 with prob and 1 with (1-prob)

      return 

    def generate_maze(self,N,S,F):

      self.N = N
      self.S = S
      self.F = F

      x, y = S[0], S[1]
      frontiers = self.get_frontier_cells(x, y, N)  #get the frontier cells of the starting point
 
      while frontiers:

        random_frontier = random.choice(tuple(frontiers))  #choose a random frontier cell
        x,y = random_frontier
        neighbours_found = self.getNeighbours(x, y, N)   
        random_neighbour = random.choice(tuple(neighbours_found))  #choose a random neighbour of the frontier cell
        z, q = random_neighbour
        self.connect_path(x,y,z,q)  #connect the path between the neighbour and the frontier cell
        frontiers.remove(tuple(random_frontier)) #remove the used frontier cell from the set
        next_frontier = self.get_frontier_cells(x, y, N)   #get the new frontiers
        for item in next_frontier:
          frontiers.add(item)   #and add them to the set
                     

    def get_frontier_cells(self,x,y,N):  #add the frontier cell of the point given that are walls
      frontier = set()
      if x in range(0,N) and y in range(0,N):
        for i, j in (x-2,y), (x+2,y), (x, y-2), (x, y+2):
          if i in range(N) and j in range(N):
            if self.grid[i,j] == 1:
              frontier.add((i,j))
      return frontier


    def getNeighbours(self,x,y,N):  #get the distance 2 neighbours that are part of the labyrinth
      neighbours = set()
      if x in range(0,N) and y in range(0,N):
        for i, j in (x-2,y), (x+2,y), (x, y-2), (x, y+2):
          if i in range(N) and j in range(N):
            if self.grid[i,j] == 0:
              neighbours.add((i,j))
      return neighbours


    def connect_path(self,x,y,z,q):   #set the frontier cell and the cell in between the neighbour and the frontier as parts of the labyrinth
      self.grid[int((x+z)/2)][int((y+q)/2)] = 0
      self.grid[x][y] = 0
              

    def draw_map(self, S=None, F=None, path=None):
        image = np.zeros((self.N, self.N, 3), dtype=int)
        image[self.grid == 0] = [255, 255, 255]
        image[self.grid == 1] = [0, 0, 0]
        if S:
            image[S] = [50, 168, 64]
        if F:
            image[F] = [168, 50, 50]
        if path:
            for n in path[1:-1]:
                image[n] = [66, 221, 245]

        plt.imshow(image)
        plt.xticks([])
        plt.yticks([])
        plt.show()
````

## Μέρος 2

Στο δεύτερο μέρος του θέματος καλούμαστε να υλοποιήσουμε διάφορους αλγορίθμους αναζήτησης για την εύρεση μονοπατιών μεταξύ δυο κόμβων και να τους εφαρμόσουμε στους λαβυρίνθους του 1ου Μέρους. Για την εφαρμογή των αλγορίθμων θα κατασκευάσουμε διάφορες συναρτήσεις κόστους τόσο για την μέτρηση του πραγματικού όσο και για την εκτίμηση των αποστάσεων από έναν κόμβο στον κόμβο στόχο (heuristic).

Οι αλγόριθμοι αναζήτησης που θα υλοποιήσουμε θα στηριχτούν στον αλγόριθμο Α*, η γενική μορφή συνάρτησης κόστους του οποίου είναι:
$$f(n)=g(n)+h(n)$$
Η συνάρτηση $g(n)$ δίνει την πραγματική απόσταση από το σημείο εκκίνησης μέχρι τον κόμβο $n$, ενώ η συνάρτηση $h(n)$ δίνει μια ευριστική εκτίμηση της απόστασης από τον κόμβο $n$ μέχρι τον στόχο. Μας ζητείται να πειραματιστούμε με τις εξής επιλογές για τις δύο συναρτήσεις:
* $g(n)=0$ και $h(n)=\{\text{manhattan}(n),\text{euclidean}(n)\}$. 
* $g(n)=1$ και $h(n)=0$. 
* $g(n)=1$ και $h(n)=\{\text{manhattan}(n),\text{euclidean}(n)\}$.

Κατασκευάσαμε τον αλγόριθμο Α* και τροποποιώντας τον κατάλληλα προσθέσαμε τις μεταβλητές $function$, $real$_$dist$ ως arguments, οι οποίες μας επιτρέπουν να τρέχουμε τον αλγόριθμο Α*, για κάθε μία από τις παραπάνω περιπτώσεις της συνάρτησης κόστους, ξεχωριστά. 

Πιο συγκεκριμένα:
* $real\_dist = False$ και $function$ = $\{$$manhattan$$\_$$dist$,$euclidean$$\_$$dist$$\}$ (περίπτωση 1η)
* $real\_dist = True$ και $function$ = $no$__$heuristic$ (περίπτωση 2η)
*  $real\_dist = True$ και $function$ = $\{manhattan\_dist,euclidean$_$dist\}$ (περίπτωση 3η)

Παρατηρούμε πως στην 1η περίπτωση, δηλαδή όταν στο κόστος $f(n)$ συμπεριλαμβάνεται μόνο η $h(n)$, προκύπτει ο αλγόριθμος αναζήτησης $Best First$, ο οποίος δεν μπορεί να βρίσκει σε κάθε περίπτωση το βέλτιστο μονοπάτι καθώς λαμβάνει υπόψιν του μόνο τις ευριστικές αποστάσεις και όχι τις πραγματικές.

Στην 2η περίπτωση,  δηλαδή όταν στο κόστος $f(n)$ συμπεριλαμβάνεται μόνο η $g(n)$, προκύπτει ο αλγόριθμος αναζήτησης $Dijkstra$, ο οποίος μας εγγυάται πως σε κάθε περίπτωση θα βρίσκει το βέλτιστο μονοπάτι.

Όσον αφορά την 3η περίπτωση που το $g(n) = 1$ και το $h(n)$ είναι μία ευριστική συνάρτηση τότε προκύπτει ο αλγόριθμος αναζήτησης Α* και διακρίνουμε 3 περιπτώσεις:
*  Εάν η $h(n)$ είναι μικρότερη (ή και ίση) από το κόστος του cell n μέχρι το σημείο F, τότε ο Α* μας εγγυάται πως θα βρεί το βέλτιστο μονοπάτι. Όσο μικρότερη είναι η τιμή του $h(n)$ τόσο ο Α* επεκτείνεται, άρα γίνεται και πιο αργός. 
* Εάν η $h(n)$ είναι αυστηρά ίση με το κόστος του cell n μέχρι το σημείο F, τότε ο Α* θα ακολουθήσει το βέλτιστο μονοπάτι και δεν θα επεκταθεί σε άλλα cells, κάνοντάς τον αρκετά γρήγορο.
*  Εάν η $h(n)$ είναι μεγαλύτερη από το κόστος του cell n μέχρι το σημείο F, τότε ο Α* δεν μας εγγυάται πως θα βρεί το βέλτιστο μονοπάτι, παρόλ' αυτά είναι γρηγορότερος.

Όσον αφορά την $Manhattan\  distance$ στην γενική της μορφή επιστρέφει:
$$ D*[abs(node[0]-end[0]) + abs(node[1]-end[1])] $$
Στην δική μας περίπτωση επιλέξαμε D = 1 (όπου 1 είναι η μικρότερη απόσταση μεταξύ γειτονικών κελιών), ώστε να πάρουμε το βέλτιστο δυνατό μονοπάτι και κατά συνέπεια η ευριστική μας συνάρτηση να είναι "admissible".

Όσον αφορά την $Euclidean\  distance$ στην γενική της μορφή επιστρέφει:
$$ D*\sqrt{(node[0]-end[0])^2 + (node[1]-end[1])^2} $$
Όμοια με πριν επιλέξαμε D=1. Επειδή η $Euclidean\ distance$ είναι μικρότερη από την $Manhattan$ θα συνεχίσουμε να παίρνουμε βέλτιστα μονοπάτια, αλλά ο Α* σε αυτήν τη περίπτωση θα είναι πιο αργός.

Τέλος μελετήθηκαν και οι επόμενες 2 ευριστικές συναρτήσεις:
$$ abs(max((end[0]-node[0]), (end[1]-node[1])))$$

$$ abs(min((end[0]-node[0]), (end[1]-node[1])))$$



### Implementation of pathfinder class

>**Important Disclaimer** :
* Τα ορίσματα c, h  της init δεν χρησιμοποιήθηκαν στο κώδικα μας καθώς περνάμε σαν όρισμα στην astar_search την συνάρτηση υπολογισμού της ευριστικής
* Δεν γράψαμε τον κώδικα μας στην def find_path(self). Αντ' αυτού φτιάξαμε μία νέα συνάρτηση (astar_search) η οποία κάνει ολή την δουλεία που χρειαζόμαστε και χρησιμοποιείται και σε αυτό το μέρος αλλά και στο μέρος 3 με την κατάλληλη χρήση των παραμέτρων.
* Όλες οι παράμετροι και γενικά η λειτουργία της συνάρτησης εξηγούνται περεταίρω με σχόλια δίπλα απο τον κώδικα


````
# from visualization import *
import heapq
import matplotlib
matplotlib.rcParams['animation.embed_limit'] = 2**128

class pathfinder:
    def __init__(self, S, F, grid, c, h, part3 = False, limits_part3 = (0,1)):

      #part3, limits_part3 are explained in part1
      #c,h are not used

        self.S = S
        self.F = F
        self.grid = grid
        self.cost = c
        self.heuristic = h
        self.part3 = part3
        self.limits_part3 = limits_part3
        #for visualization 
        self.vis = visualization(S, F)
        self.path=[]
        self.parent = None

    def find_path(self):
        
        pass

    #THIS IS THE MAIN FUNCTION OF THE CLASS
    #we implemented the code needed in here

    def astar_search(self, N, grid, start, end, function, real_dist = False, distr = None):   #false if g = 0 and true if g = 1
      
      #variable function : is used to pass as an argument each heuristic function
      #variable real_dist : if real_dist = True then g = 1 else g = 0
      #variable distr : if part3 = True then we calculate the g(n) from the distribution specified by distr

      heap = []  
      walls = []
      expanded_nodes = set()  #all nodes seen by the algorithm are saved in this set
      g_fun = {start: 0}  #dictionary to keep track of the real cost for each node
      f_fun = {start: function(self.S, self.F)}  #dictionary to keep track of the total f function for each node
      
      parent = {}   #dictionary used to keep track of the parent of each node

      heapq.heappush(heap, (f_fun[start], start))   #build the priority queue

     
      uniform_weight, triangular_weight, beta_weight = grid.distributions(N, S, self.part3, self.limits_part3)
  
      while heap:

        current_node = heapq.heappop(heap)[1]    #priority queue pops the element with the lowest total cost (f)
        expanded_nodes.add(current_node)

        if (current_node == end):  #if we find our goal
          self.path = []
          while current_node in parent:  #search backwards to track all the parents of the current node
            self.path.append(current_node)
            current_node = parent[current_node]
          #print(self.path)
          self.path.reverse()   #reverse the path
          if self.part3 == False:   #dont print if we are in the third part
            print('Length of the path found is: ',len(self.path), '\n')
            print('Number of expanded nodes is: ', len(expanded_nodes))
          self.path.insert(0,S)
          return self.path, len(expanded_nodes)   #return the path and the length of the expanded nodes
        
        for neighbour in grid.neighbours(N, current_node):   #for each neighbour of the current node found
          if grid.grid[neighbour] == 1:
            walls.append(neighbour)
          else:                         #if its not a wall
            if self.part3 == False:
              
              if real_dist == True:   #if we arent at part3 and real cost isnt zero
                g_score = g_fun[current_node] + 1      #increase the real cost of the current node by one
              else:
                g_score = 0       #else g is zero every time
            else:                 #if part3 is True them calculate the real cost by adding to the real cost of the current node 
              if distr == 'uniform':      #the value generated from the distribution needed
                g_score = g_fun[current_node] + uniform_weight[neighbour]
              elif distr == 'beta':
                g_score = g_fun[current_node] + beta_weight[neighbour]
              elif distr == 'triangular':
                g_score = g_fun[current_node] + triangular_weight[neighbour]
              else:
                raise NameError('You should  give a viable name for the distribution (uniform, beta, triangular)')

            #if the neighbour is already visited and the total cost found this time is greater or equal to the previous cost of the same neighbour
            #do nothing 
            if neighbour in expanded_nodes and (function(neighbour,self.F) + g_score)>= f_fun.get(neighbour,0):
              continue
            #if the neighbour isnt in expanded_noded or the total cost found this time is lesser than the previous cost of the same neighbour
            if (function(neighbour,self.F) + g_score)< f_fun.get(neighbour,0) or neighbour not in [i[1] for i in heap]: 
              parent[neighbour] = current_node   #update the parent of the neighbour
              g_fun[neighbour] = g_score         #update the real cost of the neighbour
              f_fun[neighbour] = function(neighbour, self.F) + g_score    #update the total cost of the neighbour
              heapq.heappush(heap, (f_fun[neighbour], neighbour))         #push the neighbour in the priority queue
      
        #CHANGED DRAW STEP : {for (_, (x,y)) in frontier:}
        self.vis.draw_step(grid,heap,expanded_nodes)    
 
      return False



    def euclidean_dist(self, node, end):  #case g=0 and h={euclidean}
      return  np.sqrt((end[0]-node[0])**2 + (end[1]-node[1])**2)

    def manhattan_dist(self, node, end):  #case g=0 and h={manhattan}
      return abs(node[0]-end[0]) + abs(node[1]-end[1])
    
    def no_heuristic(self,node,end):  #case h=0 (g is to be determined)
      return 0 

    #extra cases
    
    def max_heuristic(self, node, end):   
      return abs(max((end[0]-node[0]), (end[1]-node[1])))  #return abs(max(Δx, Δy))

    def min_heuristic(self, node, end):
      return abs(min((end[0]-node[0]), (end[1]-node[1])))  #return abs(min(Δx, Δy))

    def get_path(self):
        return self.path

````

````
## YOUR CODE HERE

def plot_ditributions(up_low_limit):

  a,b = up_low_limit
  distributions = ['uniform', 'triangular', 'beta']
  
  #initialize the lists used to store the information

  path_len_euclidean = [None] * 10
  path_len_manhattan = [None] * 10
  path_len_no_heuristic = [None] * 10
  path_len_min = [None] * 10
  path_len_max = [None] * 10

  expanded_nodes_euclidean = [None] * 10
  expanded_nodes_manhattan = [None] * 10
  expanded_nodes_no_heuristic = [None] * 10
  expanded_nodes_min = [None] * 10
  expanded_nodes_max = [None] * 10


  S = (1,3)  #starting point

  for distribution in distributions:
    i = 0
    for N in range(10,110,10):  #size of the grid in [10,100] with a step of 10
      F = (N-3, N-4)     #ending point with respect to N
      
      map = grid(N, S, F, True, up_low_limit)
      pf = pathfinder(S, F, map, lambda x, y: 1, lambda x, y: 0, True, up_low_limit)

      path, expanded_nodes_manhattan[i] = pf.astar_search(N, pf.grid,S,F, pf.manhattan_dist, False, distribution)
      path_len_manhattan[i] = len(path) -1  #start excluded

      path, expanded_nodes_euclidean[i] = pf.astar_search(N, pf.grid,S,F, pf.euclidean_dist, False, distribution)
      path_len_euclidean[i] = len(path) -1  

      path, expanded_nodes_no_heuristic[i] = pf.astar_search(N, pf.grid,S,F, pf.no_heuristic, False, distribution)
      path_len_no_heuristic[i] = len(path) -1  

      path, expanded_nodes_min[i] = pf.astar_search(N, pf.grid,S,F, pf.min_heuristic, False, distribution)
      path_len_min[i] = len(path) -1  

      path, expanded_nodes_max[i] = pf.astar_search(N, pf.grid,S,F, pf.max_heuristic, False, distribution)
      path_len_max[i] = len(path) -1  
    
      i += 1
    
    
    f, (ax1, ax2) = plt.subplots(1, 2, figsize = (15,6))
    plt.suptitle('{} distribution in {}'.format(distribution, up_low_limit), fontsize = 15)
    #first subplot
    x = np.arange(10,110,10)
    ax1.plot(x, [q for q in path_len_manhattan], label = 'manhattan')
    ax1.plot(x, [q for q in path_len_euclidean], label = 'euclidean')
    ax1.plot(x, [q for q in path_len_no_heuristic], label = 'no_heuristic')
    ax1.plot(x, [q for q in path_len_min], label = 'min')
    ax1.plot(x, [q for q in path_len_max], label = 'max')
    ax1.set_title('Length of the path found using {} distribution'.format(distribution))
    ax1.set_xlabel('Grid size')
    ax1.set_ylabel('Length of the path')
    ax1.legend()
  
    #second subplot
    ax2.plot(x, [q for q in expanded_nodes_manhattan], label = 'manhattan')
    ax2.plot(x, [q for q in expanded_nodes_euclidean], label = 'euclidean')
    ax2.plot(x, [q for q in expanded_nodes_no_heuristic], label = 'no_heuristic')
    ax2.plot(x, [q for q in expanded_nodes_min], label = 'min')
    ax2.plot(x, [q for q in expanded_nodes_max], label = 'max')
    ax2.set_title('Number of expanded nodes found using {} distribution'.format(distribution))
    ax2.set_xlabel('Grid size')
    ax2.set_ylabel('# of expanded nodes')
    ax2.legend()



````

Για την καλύτερη κατανόηση των συμπερασμάτων θα τα χωρίσουμε σε sections ανάλογα τις τιμές που δίνουν οι κατανομές.

>**Αποτελέσματα των κατανομών στο διάστημα (0,1):**

Επειδή το πραγματικό κόστος ανήκει στο διάστημα (0,1) πλέον η ευριστική αποτελεί σημαντικότερο παράγοντα της συνολικής συνάρτησης κόστους και κατά συνέπεια καθορίζει με μεγαλύτερο "βάρος" την επιλογή του μονοπατιού. Όπως παρατηρούμε απο τις γραφικές η κάθε ευριστική βρίσκει διαφορετικό μήκος για το μονοπάτι πράγμα το οποίο σημαίνει ότι δεν εντοπίζουμε πάντα το βέλτιστο μονοπάτι και κατα συνέπεια επιβεβαιώνεται ο ισχυρισμός μας. Εκτός απο το διάστημα στο οποίο η πραγματική απόσταση παίρνει τιμές σημασία φαίνεται να έχει και η επιλογή της εκάστοτε κατανομής η οποία θα απονείμει αυτές τις τιμές. Παρατηρούμε:
* Στην **uniform distribution** λόγω της ομοιομορφίας της κατανομής παρατηρούμε ότι οι γραφικές ακολουθούν περίπου την ίδια πορεία με μικρές διαφοροποιήσεις οι οποίες οφείλονται στην μορφή της ευριστικής συνάρτησης
* Στην **triangular distribution** παρατηρούνται περισσότερα spikes στις γραφικές 
* Στην **beta distribution** σε μικρότερα grid οι διαφορές μεταξύ των γραφικών είναι ελάχιστες, ενώ όσο μεγαλώνει ο αριθμός του λαβυρίνθου φαίνεται να υπάρχει μεγαλύτερη απόκλιση. Επίσης σε αυτή την κατανομή παρατηρούμε τον μεγαλύτερο αριθμό απο expanded nodes (Dijkstra).

>**Αποτελέσματα των κατανομών στο διάστημα (1,2):**

Πλέον το πραγματικό κόστος αποτελεί κύριο παράγοντα για την εύρεση του συνολικού κόστους με αποτέλεσμα όλες οι γραφικές να έχουν σχεδόν την ίδια μορφή. Αυτό οφείλεται στο γεγονός ότι η διαφορετική ευριστική για κάθε περίπτωση δεν επηρεάζει καθοριστικά την απόφαση του αλγορίθμου και κατά συνέπεια η διαφοροποίηση που θα έδινε η κάθε ευριστική δεν παρατηρείται.

>**Αποτελέσματα των κατανομών στο διάστημα [0,0.1):**

Σε αυτή την περίπτωση οι ευριστικές μας δεν είναι admissible αφού ισχύει ότι:
$$ h(n) > g'(n)\ όπου:$$

* n είναι ο κάθε κόμβος
* g'(n) είναι η πραγματική απόσταση του κάθε κόμβου απο τον τελικό στόχο

Επομένως οι γραφικές για το μονοπάτι πλέον έχουν αρκετή απόκλιση η μία απο την άλλη, πράγμα το οποίο είναι λογικό αφού ακολουθούν διαφορετικές ευριστικές. Τέλος παρατηρούμε ότι σταθερά το χειρότερο μονοπάτι εντοπίζεται χρησιμοποιώντας την min_heuristic.
